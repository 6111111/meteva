```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import meteva.base as meb
import meteva.method as mem
import meteva.product as mpd
import numpy as np
import datetime
import copy
import matplotlib.pyplot as plt
import time
import pandas as pd
import math
import matplotlib as mpl
```

```python
station_path = r"\\10.20.90.100\GridDataShare100\jianyan\ENSEMBLE\National_OBS\BT19022020.000"
station = meb.read_station(station_path)
```

```python
times = datetime.datetime(2019,3,1,8,0)
timee = datetime.datetime(2019,3,4,20,0)
```

```python
#读取观测
dir_ob = r"\\10.20.90.100\GridDataShare100\jianyan\ENSEMBLE\National_OBS\BTYYMMDDHH.000"
time0 = times
ob_list = []
while time0<= timee:
    path = meb.tool.path_tools.get_path(dir_ob,time0)
    sta = meb.read_stadata_from_micaps3(path,station = station)
    if sta is not None:
        meb.set_stadata_coords(sta,time = time0)
        ob_list.append(sta)
    time0 = time0 + datetime.timedelta(hours = 12)
ob_all = pd.concat(ob_list,axis = 0)
meb.set_stadata_names(ob_all,["ob"])
```

```python
#读取ec_mean数据
dir_ec = r"\\10.20.90.100\GridDataShare100\jianyan\ENSEMBLE\YYYYMMDDHH\EC_Mean_r24.nc"
ec_list = []
time0 = times 
while time0 <=timee:
    time_ut = time0 + datetime.timedelta(hours = -8)
    path = meb.tool.path_tools.get_path(dir_ec,time_ut)
    grd = meb.read_griddata_from_nc(path)
    if grd is not None:
        sta = meb.interp_gs_linear(grd,station)
        meb.set_stadata_coords(sta,time = time0)
        ec_list.append(sta)
    time0 = time0 + datetime.timedelta(hours = 12)
ec_all = pd.concat(ec_list,axis = 0)
meb.set_stadata_names(ec_all,["ec_mean"])
```

```python
#读取集合预报数据
dir_ec = r"\\10.20.90.100\GridDataShare100\jianyan\ENSEMBLE\YYYYMMDDHH\EC_Members_r24.nc"
ec_member_list = []
time0 = times 
while time0 <=timee:
    time_ut = time0 + datetime.timedelta(hours = -8)
    path = meb.tool.path_tools.get_path(dir_ec,time_ut)
    grd = meb.read_griddata_from_nc(path)
    if grd is not None:
        sta = meb.interp_gs_linear(grd,station)
        meb.set_stadata_coords(sta,time = time0)
        ec_member_list.append(sta)
    time0 = time0 + datetime.timedelta(hours = 12)
ec_member_all = pd.concat(ec_member_list,axis = 0)
```
