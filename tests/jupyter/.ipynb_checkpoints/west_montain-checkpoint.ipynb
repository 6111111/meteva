```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import meteva.base as meb
import meteva.method as mem
import meteva.product as mpd
import numpy as np
import datetime
import copy
import matplotlib.pyplot as plt
import time
import pandas as pd
import math
import matplotlib as mpl
import os
```

```python
dir_nc = r"L:\wuhan_data\nc\YYYYMMDD\hbnw.tHH00Z.wrfprs"
dir_micaps_rain = r"L:\wuhan_data\micaps\rain01\YYMMDDHH.TTT"
dir_png = r"L:\wuhan_data\png\rain01\YYMMDDHH.TTT.png"
times = datetime.datetime(2019,6,5,0,0)
timee = datetime.datetime(2019,6,22,0,0)
grd0 = meb.read_griddata_from_nc(r"L:\wuhan_data\nc\20190605\hbnw.t0000Z.wrfprs60_surface2.nc",value_name="tp")
grid0 = meb.get_grid_of_data(grd0)
clev1,cmap1 = meb.tool.color_tools.clev_cmap_rain_1h()
```

```python
time0 = times
timee= time0
while time0 <= timee:
    dir1 = meb.tool.path_tools.get_path(dir_nc,time0)
    grd0 = meb.grid_data(grid0)
    for dh in range(1,13):
        dm = dh * 60
        path_nc = dir1 + str(dm) + "_surface2.nc"
        print(path_nc)
        if os.path.exists(path_nc):
            path_m4 = meb.tool.path_tools.get_path(dir_micaps_rain,time0,dh)
            grd_r = meb.read_griddata_from_nc(path_nc,value_name="tp")
            grd1 = meb.grid_data(grid0)
            grd1.values = grd_r.values - grd0.values
            meb.set_griddata_coords(grd1,gtime = [time0],dtime_list = [dh],member_list = ["hbnw"])
            grd1.name = "rain01"
            meb.write_griddata_to_micaps4(grd1,path_m4,creat_dir=True,effectiveNum=3)
            path_png = meb.tool.path_tools.get_path(dir_png,time0,dh)
            meb.tool.plot_tools.contourf_2d_grid(grd1,clevs = clev1,cmap = cmap1,save_path = path_png)
            grd0 = grd_r
            print("a")
    time0 = time0 + datetime.timedelta(hours=1)
```
