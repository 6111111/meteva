```python
nam1_path = r"H:\resource\meteva\sta\station_name1.txt"
file1 = open(nam1_path)
str0 = file1.read()
file1.close()
str1 = str0.split("\n")
print(len(str1))
for i in range(len(str1)):
    strs_line = str1[i].split("\t")
    if len(strs_line) ==3:
        values = strs_line[0]
        strs_int = ""
        for i in range(len(values)):
            strs = values[i]
            for s in strs:
                if s.isdigit():
                    strs_int += s
                else:
                    strs_int += str(ord(s))
        print(strs_int)
        key =  int(strs_int)
        if not key in id_name_dict.keys():
            value1 = strs_line[1]+"_"+strs_line[2]
            id_name_dict[key] = value1
print(len(id_name_dict.keys()))
```
