```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import meteva.base as meb
import pandas as pd
import numpy as np
import datetime
import copy
```

```python
data = {"id":[54511,54522],"lon":[100,110],"lat":[30,40],"ob":[0,0],"grapes":[0,0],"ec":[0,0],}
df = pd.DataFrame(data)
sta = meb.sta_data(df)
meb.set_stadata_coords(sta,level = 1000,time = datetime.datetime(2019,12,31,8,0),dtime = 0)
sta1 = copy.deepcopy(sta)
meb.set_stadata_coords(sta1,level = 850)
sta = meb.combine_join(sta,sta1)
sta1 = copy.deepcopy(sta)
meb.set_stadata_coords(sta1,time = datetime.datetime(2020,1,1,20,0))
sta = meb.combine_join(sta,sta1)
sta1 = copy.deepcopy(sta)
meb.set_stadata_coords(sta1,dtime = 24)
sta_all = meb.combine_join(sta,sta1)
sta_all.iloc[:,-3:] = (np.random.rand(16,3)*100).astype(np.int16)
sta_all.iloc[10:13,-1] = meb.IV
print(sta_all)
```

#  将数据分组   
**group((data,g = None,gll = None)**   
将站点数据按指定方式进行分组  


**参数说明：**  
  **<font face="黑体" color=green size=5>data</font>**  ：[站点数据](https://www.showdoc.cc/nmc?page_id=3744334022014027)  
  **<font face="黑体" color=green size=5>g</font>** ：指定分组维度，对预报观测样本进行分组检验时的分组依据，缺省是为None，即不进行分组检验，非缺省时为字符串形式，可选项包括  

|可选项| 说明|  
|:--|:--|  
|None| 不分组，返回原数据和gll的原值|  
|"level" |      按层次进行分组|    
|"time"|       按起报时间分组|   
|"time_range"|   按起报时间区间分组|   
|"year"|       按起报时间中的年份进行分组|  
|"month"|       按起报时间中的月份进行分组|  
|"day"|         按起报时间中的日期进行分组|  
|"dayofyear"|     按起报时间中的日期在一年中的排序（day_of_year）进行分组|  
|"hour"|       按起报时间的小时数（hour_of_day）进行分组|  
|"ob_time"|     按观测时间（起报时间+预报时效）进行分组|  
     "ob_time_range"| 按观测时间（起报时间+预报时效）区间进行分组|  
|"ob_year"|     按观测时间（起报时间+预报时效）中的年份进行分组|  
|"ob_month"|     按观测时间（起报时间+预报时效）中的月份进行分组|  
|"ob_day"|      按观测时间（起报时间+预报时效）中的日期进行分组|
|"ob_dayofyear"|  按观测时间（起报时间+预报时效）中的日期在一年中的排序（day_of_year）进行分组|   
|"ob_hour"|      按观测时间（起报时间+预报时效）中的小时数（hour_of_day）进行分组|     
|"dtime"|        按预报时效进行分组|  
|"dtime_range"|   按预报时效区间进行分组|  
|"dday"|        按预报时效包含的天数进行分组,例如预报时效为0、12、24和36小时的样本会分别被标记为 0、1、1和2|  
|"dhour"|        按预报时效整除24小时后的余数进行分组,例如预报时效为12、24和36小时的样本会分别被标记为 12,0和12|  
|"id"|          按站号进行分组|  
|"lon_range"|     按经度区间分组|  
|"lon_step"|      按经度进行等间距分组|  
|"lat_range"|     按维度区间分组|  
|"lat_step"|      按纬度进行等间距分组|  
|"last_range"|     按最后一列数据值的取值区间进行分组,分组后最后一列数据将被丢弃|  
|"last_step"|     根据最后一列数据值的大小进行等距分组,分组后最后一列数据将被丢弃| 
|"last_step"|     按站号进行分组| 

  **<font face="黑体" color=green size=5>gll**: 指定分组的取值区间  
  
  |适用情况|参数形式|示例及效果|
|:---|:---|:---|
|g="lon_step", "lat_step"或 "last_step"时| 单层列表，其中第0个元素为等距分割的起始值，第1个元素为等级分割的间隔| 例如当站点数据集分布在经度范围为69-135°E的区域内，如果g="lon",gll = [100,10],则会将所有的数据按照[60,70),[70,80),[80,90),[90,100),[100,110),[110,120),[120,130)[130,140)共8个经度区间（前闭后开）分组|
|g="time_range", "ob_time_range", "dtime_range", "lon_range", "lat_range"或 "last_range"时| 双层列表，其中每一个内层列表的第0个元素为区间的起始值，第1个元素为区间的结束值（闭区间），不同区间可重叠| 例如当站点数据集分布在经度范围为69-139°E的区域内，如果g = "lon",gll = [[80,90],[110-120]],则会从数据中提取经度位于80-90°E和110-120°E的两组数据|
|g="grid"时|单层列表，列表中的每个元素为一个[网格信息类](https://www.showdoc.cc/meteva?page_id=3975600815874861)变量|例如当g = "grid", gll = [meb.grid([100,110,1],[20,30,1]),  meb.grid([120,130,1],[40,50,1])]时，就可以提取位于西南南部和东北的两块矩形区域的数据，并分为2组|  
|g等于其它可选项时|None|当g!=None时，按g参数对应的数据取值相同的分为一组，分组数量和数据取值的集合size一致，例如原始数据中包含54511, 54512, 54513三个站点，则会按站点号将所有相同站号的数据划分为一组，一共分为3组|    
|同上|双层列表，其中内层列表的长度可以不固定，不同列表可重叠,如果内层列表size = 1,也可以不用列表形式，直接列出其中内容|按g参数对应的数据，根据内层列表所列值进行分组例如原始数据中包含54511, 54512, 54513三个站点，如果gll ==[54511,[54511,54512],[54512,54513]] 则分为3组,第一组包含54511站,第2组包含所有站号为54511和54512的数据，第3组包含所有54512和54513。   |
  
  
  **<font face="黑体" color=blue size=5>return**</font>：  有效的分组结果和分组方式，python元组数据，其中第一个元素为列表，列表中的元素为每个分组后的[站点数据](https://www.showdoc.cc/nmc?page_id=3744334022014027)，元组的第二个元素为实际分类所采用的gll。在分组时，如果gll指定的分组里没有包含任何站点，则该分组参数会从gll中移除。

**调用示例**

```python
result = meb.group(sta_all)  
print(result[0])
print(result[1])#不分组，返回值的第一个元素为站点数据，第二个元素为None
```

```python
result = meb.group(sta_all,g ="time")  #group_list_list = None
print(result[0]) 
print(result[1]) #按起报时间分组，返回的第一个元素为两组站点数据构成的列表，第二个元素实际采用的分组方式
```

```python
result = meb.group(sta_all,g ="time",gll = [["2019123108","2020010120"],"2020010120","2030010108"])  
print(result[0])#按指定方式的时间分组，分组结果中第一组包含了两天的数据样本，第二组包含了一天的数据样本
print(result[1]) #第三组没有包含数据样本，所以在结果中被删除了。
```

```python
result = meb.group(sta_all,g ="time_range",gll = [["2019123108","2020010108"],["2020010120","2020010220"]])  
print(result[0])#按指定方式的时间分组，每一组根据列表给出的起止时间选取相应的数据，一共分成了2组
print(result[1]) 
```

```python
grid1 = meb.grid([95,105,1],[25,35,1])
grid2 = meb.grid([105,115,1],[35,45,1])
result = meb.group(sta_all,g ="grid",gll = [grid1,grid2])
print(result[0])#按指定网格区域，一共分成了2组
print(result[1]) 
```

```python
result = meb.group(sta_all,g = "lat_step",gll = [30,5])
print(result[0])#按指定网格区域，一共分成了2组
print(result[1]) 
```

```python
sta_all["div"] = (np.random.randn(16) * 100).astype(np.int16)  
print(sta_all) # 增加一列数据作为分组的依据，它可以是散点，涡度，下垫面类型等等
```

```python
result = meb.group(sta_all,g = "last_range",gll =[[-200,0],[0,200]])
print(result[0])#根据最后一列数据分组，假设div代表散度的话，这样可将数据样本分为散度为正和为负的两组样本
print(result[1]) 
```

```python
result = meb.group(sta_all,g = "last_step",gll =[0,100])
print(result[0])#根据最后一列数据分组，假设div代表散度的话，这样可将数据样本分为散度为不同的取值区间是的多组样本
print(result[1]) 
```
